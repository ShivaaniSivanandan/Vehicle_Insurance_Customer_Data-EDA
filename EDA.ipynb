### 2. Checking and Cleaning Data Quality:

### i. Null values

### Generate a summary of count of all the null values column wise

```python
customer_details.isnull().sum()
```

```python
customer_policy.isnull().sum()
```

### Drop Null values for customer_id because central tendencies for idâ€™s is not feasible.

```python
customer_details = customer_details.dropna(subset=['customer_id'])
customer_policy = customer_policy.dropna(subset=['customer_id'])
```

### Replace all null values for numeric columns by mean. 

```python
customer_details.fillna(customer_details.mean(), inplace=True)
customer_policy.fillna(customer_policy.mean(), inplace=True)
```

### Replace all null values for Categorical value by mode.

```python
customer_details.fillna(customer_details.mode(), inplace=True)
customer_policy.fillna(customer_policy.mode(), inplace=True)
```

### ii. Outliers

```python
customer_policy.describe()
```

### Generate a summary of count of all the outliers column wise

```python
dQ1 = customer_details.quantile(0.25)
dQ3 = customer_details.quantile(0.75)
dIQR = dQ3 - dQ1
print(dIQR)
```

```python
print('customer_details')
print('')
dlow = dQ1 - 1.5 * dIQR
dhigh = dQ3 + 1.5 * dIQR
print('Lower than Q1')
print(dlow)
print('')
print('Higher than Q1')
print(dhigh)
```

```python
pQ1 = customer_policy.quantile(0.25)
pQ3 = customer_policy.quantile(0.75)
pIQR = pQ3 - pQ1
print(pIQR)
```

```python
print('customer_policy_details')
print('')
plow = pQ1 - 1.5 * pIQR
phigh = pQ3 + 1.5 * pIQR
print('Lower than Q1')
print(plow)
print('')
print('Higher than Q1')
print(phigh)
```

### Replace all outlier values for numeric columns by mean.

### 3. Create a Master table for future use. Join the customer table and customer_policy table to get a master table using customer_id in both tables.
